```python
from pyspark.sql.functions import col,lower
```

```python

df = spark.table("associate_assignment.default.bronze_master_customer")
```

```python
df_valid_customer = (
    df
    .filter(
        (col("customer_id").rlike("^[0-9]+$")) &
        (col('name').isNotNull()) & (col('name') != " ") &
        (col("city").isNotNull() & lower(col("status")).isin(["active", "inactive","pending","deleted"])))
    .withColumn("customer_id", col("customer_id").try_cast("int"))
    .filter(col("customer_id") > 0)
)
```

```python
%sql
drop table if exists associate_assignment.default.dim_master_customer;
```

```python

df_valid_customer.write.mode("overwrite").saveAsTable("associate_assignment.default.dim_master_customer")
```
